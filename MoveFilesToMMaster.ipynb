## Move folders to process on MMASTER

```python
import os
import shutil

# Paths
zip_folder = r"D:\Dwnl_Arch_ASTER_CBlanca\Data"
met_folder = r"D:\Dwnl_Arch_ASTER_CBlanca\Data"
output_base = r'C:\P_MMaster\Huayhuash'

# List of base names (example: 'AST_L1A_00301042012153358')
List_to_move = [
'AST_L1A_00307112019152810',
'AST_L1A_00307172024144354',
'AST_L1A_00305252002152946',
'AST_L1A_00307112019152801',
'AST_L1A_00307172024144345',
'AST_L1A_00305252002152955',
'AST_L1A_00310202001153939',]

# Funci√≥n principal
def copy_files(base_name):
    # print(f"\nüîç Processing: {base_name}")
    
    # Buscar .zip y .zip.met que empiecen con base_name
    zip_file = next((f for f in os.listdir(zip_folder) if f.startswith(base_name) and f.endswith(".zip")), None)
    met_file = next((f for f in os.listdir(met_folder) if f.startswith(base_name) and f.endswith(".zip.met")), None)
    
    if not zip_file or not met_file:
        print(f"‚ö†Ô∏è  Files not found for {base_name}. Skipping.")
        return

    # Crear directorio destino
    target_dir = os.path.join(output_base, base_name)
    os.makedirs(target_dir, exist_ok=True)
    print(f"üìÇ Created target dir: {target_dir}")

    # Rutas completas
    src_zip = os.path.join(zip_folder, zip_file)
    src_met = os.path.join(met_folder, met_file)
    dst_zip = os.path.join(target_dir, zip_file)
    dst_met = os.path.join(target_dir, met_file)

    # Copiar archivos
    shutil.copy2(src_zip, dst_zip)
    shutil.copy2(src_met, dst_met)
    print(f"‚úÖ Copied: {zip_file} and {met_file}")

# Ejecutar para cada ID
for base_name in List_to_move:
    copy_files(base_name)

```
